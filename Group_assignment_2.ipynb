```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

```python
df = pd.DataFrame(data,columns=['match_id','city','venue','team1','team2','toss_winner','toss_decision','first_innings_total','winner',
                               'home_win_percentage','away_win_percentage','result'])
```

```python
df
```

```python
df1= df.drop(['match_id','city','team1','team2','winner'],axis=1)
```

```python
df1
```

```python
categorical_column = ['venue','toss_winner','toss_decision']
```

```python
from sklearn.preprocessing import OneHotEncoder
oneHotEncoding = OneHotEncoder(drop="first", sparse=False)
new_df = oneHotEncoding.fit_transform(df1[categorical_column])
new_df = pd.DataFrame(data=new_df,columns= oneHotEncoding.get_feature_names())
df1 = df1.drop(columns= categorical_column)
for column in new_df.columns:
    df1[column]=new_df[column]
df1.head()
```

```python
to_be_scaled = ['first_innings_total','home_win_percentage','away_win_percentage']
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df1[to_be_scaled])
scaled_data
```

```python
scaled_data = pd.DataFrame(data=scaled_data, columns= to_be_scaled)
scaled_data
```

```python
df2 = df1.drop(to_be_scaled,axis=1)
df2= df2.drop('result',axis=1)
df2
```

```python
final_df= pd.concat([df2,scaled_data,df1['result']],axis=1)
final_df
```

```python
y = final_df["result"]
X= final_df.drop(["result"],axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
clf = SVC(kernel='linear')
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python

from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf','linear','sigmoid']}
 
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)
 
# fitting the model for grid search
grid.fit(X_train, y_train)
```

```python
# print best parameter after tuning
print(grid.best_params_)
 
# print how our model looks after hyper-parameter tuning
print(grid.best_estimator_)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
grid_predictions = grid.predict(X_test)
 
# print classification report
print(classification_report(y_test, grid_predictions))
```

```python
predicted_df = pd.DataFrame(grid_predictions,columns=['predicted_team1_result'])
```
